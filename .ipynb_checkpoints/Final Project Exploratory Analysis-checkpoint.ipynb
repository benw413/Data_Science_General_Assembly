```python
at.dropna(subset= ['Hike Type'],inplace = True)
```

```python
def clean_8(df):
    to_drop = ['Section','Alternative','Othe','Oher','alternative','Alterntive']
    x = df[df['Hike Type'].isin(to_drop)]
    y = x.index
    df.drop(y,inplace = True)
clean_8(at)
```
